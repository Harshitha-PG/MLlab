```python
hearts_df.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
hearts_df.columns
```

```python
X=hearts_df.drop('target',axis=1)
y=hearts_df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression()
```

```python
classifier.fit(X_train,y_train)
```

```python
y_predicted=classifier.predict(X_test)
```

```python
y_predicted[0:5]
```

```python
from sklearn import metrics
```

```python
con_matrix=metrics.confusion_matrix(y_test,y_predicted)
con_matrix
```

```python
acc =metrics.accuracy_score(y_test,y_predicted)
prec =metrics.precision_score(y_test,y_predicted)
rec =metrics.recall_score(y_test,y_predicted)
print(f'Accuracy ={acc} \nPrecision = {prec} \nRecall ={rec}')
```

```python
plt.title("Confusion Matrix")
sns.heatmap(pd.DataFrame(con_matrix),annot = True,cmap='YlGnBu')
plt.show()
```
